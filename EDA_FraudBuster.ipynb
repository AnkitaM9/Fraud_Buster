1. Importing the Basic Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Upload Dataset

```python
data=pd.read_csv('dataset/creditcard_2023.csv')
```

3. Exploring DataFrame

```python
data.shape
data.head()
```

```python
data.info()
data.describe()
```

3.1  Checking for missing values

```python
missing_data=data.isnull().any()
print(missing_data)
```

4.  Visualization Type-Histogram 
* Distribution of dataset

```python
data.hist(bins=50, figsize=(20, 15) )
plt.show
```

5. Sperated dataset into Not Fraud and Fraud

```python
fraud=data[data['Class']==1]
not_fraud=data[data['Class']==0]
print(f"Not Fraud [0]: {not_fraud.shape}, Fraud [1]: {fraud.shape}")
```

6. Class Distribution

```python
category_counts=data['Class'].value_counts()
print(category_counts)
```

```python
plt.figure(figsize=(6, 4))
sns.catplot(x='Class', kind='count', data=data, hue='Class' , palette='Set2')
plt.title('Distribution of Class')
plt.xticks([0, 1], [' Not Fraud', 'Fraud', ])
plt.ylabel('Count') 
plt.show() 
```

6.1 Analyzing the amount of money involved in different transaction classes.

```python
not_fraud.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
sns.displot(data=data, bins=50, x='Amount', kind='hist',hue='Class', aspect=1.5)
plt.title('Transaction Amount per Class')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```
